```python
f= open('../datasets/shakespeare.txt','r')
d=str(f.read())
f.close()
```

```python
data= d.strip().lower()
data=data[104:]
```

```python
print len(data)
```

```python
order=1
lm={}
n=100000
for ix in range(n):
    dt= '*'*order+data
    ctx= dt[ix:ix+order]
    nxt=dt[ix+order]
    
    #print ctx, nxt
    try:
        lm[ctx][nxt]+=1
    except:
        try:
            lm[ctx][nxt]=1
        except:
            lm[ctx]={}
            lm[ctx][nxt]=1
        
```
